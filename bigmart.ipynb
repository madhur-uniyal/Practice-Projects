## preprocessing the dataset

```python
# checking the null values in a dataset
mart.isnull().sum()
```

```python
# printing the categorical columns
for col in mart.select_dtypes(include=[object]):
    print(mart[col].value_counts())
    print("\n")
```

```python
# filling the missing values of the column"Item_Weight" by their mean
mart["Item_Weight"]=mart["Item_Weight"].fillna(mart["Item_Weight"].mean())
```

```python
mart
```

```python
#checking again the null values
mart.isnull().sum()
```

```python
# filling the missing values of the column 'Outlet_Size' by their mode(as its is categorical data)
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='most_frequent')
mart['Outlet_Size']=imp.fit_transform(mart['Outlet_Size'].values.reshape(-1,1))
mart
```

```python
mart.isnull().sum()
```

```python
#Now if we see in the dataset the item visibility column some values are zero for corresponding item
#so we need to replace it by the mean of the that column. 
mart["Item_Visibility"]=mart["Item_Visibility"].replace(0,mart["Item_Visibility"].mean())
```

```python
mart
```

```python
#now while printing categorical columns above we have seen that in column "item fat content" same fat content is represented
#in different ways like Lower fat or LF or low fat and regular or reg.So we need to add them to bring in same category
mart["Item_Fat_Content"]=mart["Item_Fat_Content"].replace({"LF":"Low Fat","reg":"Regular","low fat":"Low Fat"})
```

```python
# now data is properly categorized in Low Fat and regular
mart["Item_Fat_Content"].value_counts()
```

### Label Encoding

```python
# now we will convert catergorical attributes to numerical
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_col=["Item_Identifier","Item_Fat_Content","Item_Type","Outlet_Identifier","Outlet_Size","Outlet_Location_Type","Outlet_Type","new_item"]
for col in cat_col:
    mart[col]=le.fit_transform(mart[col])
```

```python
mart = pd.get_dummies(mart, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type',
                              'Outlet_Identifier'])
```

### Input split

```python
# now before training the model lets check the skewness and outlier also
mart.skew()
```

```python
#before removing the skewness, we checking for the outliners in the dataset.And if any outliner is present then its removed with help of zscore
from scipy.stats import zscore
z_score=abs(zscore(mart))
print(mart.shape)
df_martx=mart[(z_score<3).all(axis=1)]
print(df_martx.shape)
```

```python
# we will now split the data and we will remove the item_identifier column as we have made new_item column from that only
df_x=df_martx.drop(columns=["Item_Identifier","Item_Outlet_Sales","Outlet_Establishment_Year"],axis=1)
y=df_martx["Item_Outlet_Sales"]
```

```python
#After removing outliner we need to remove skewness which is done by boxcox
from scipy.stats import boxcox
for col in df_martx.columns:
    if df_martx.skew().loc[col]>0.55:
        df_martx[col]=np.log1p(df_martx[col])
```

```python
df_martx.skew()
```

```python
#still Item_Visibility column has skewness>0.55 so we can also implement squareroot techniquw ro remove skewed data 
df_martx["Item_Visibility"]=np.sqrt(df_martx["Item_Visibility"])
```

```python
#skewness is removed from all columns.Now we can go for training the model
df_martx.skew()
```

```python
#before training the model we need to put data in one standard using standardScaler technique
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_martx)
x=pd.DataFrame(data=x,columns=df_martx.columns)
x
```

## Model Training

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state = 1)
```

```python
x.shape
```

```python
y.shape
```

```python
#now for model training we will import some libraries
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

def train(m,x,y):
    #train the model
    m.fit(x_train,y_train)
    
    #predict the model
    pred=m.predict(x_test)
    
    #perform cross validation
    cv_score=cross_val_score(m,x,y,scoring="r2")
    
    print("Model Report")
    print("Mean squared error",mean_squared_error(y_test,pred))
    print(' root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
    print("r2 score",r2_score(y_test,pred))
    print("CV score",cv_score.mean())
```

```python
#using model linear regression as target variable is numeric 
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
train(lr,x,y)
```

```python
#using regularization technique so to solve over fitting if any
from sklearn.linear_model import Lasso,Ridge
```

```python
#lasso technique
ls=Lasso(alpha=0.0001)
train(ls,x,y)
```

```python
#ridge technique
rd=Ridge(alpha=0.0001)
train(rd,x,y)
```

```python
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.0001)
train(enr,x,y)
```

```python
#ensemble technique for boosting
#RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=300,random_state=46)
train(rfr,x,y)
```

```python
#AdaBoostRegressor
from sklearn.ensemble import AdaBoostRegressor
adr=AdaBoostRegressor()
train(adr,x,y)
```

```python
# Ridge Technique is giving minimum error after using regularization to avoid overitting so this is best model
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state = 1)
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
pred=rd.predict(x_test)
print("Model Report")
print("Mean squared error    : ",mean_squared_error(y_test,pred))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
print("r2 score:               ",r2_score(y_test,pred))
```
