```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#csv file reading using pandas
df_happiness=pd.read_csv("D:/Datatrainedprojects/happiness_score_dataset.csv")
df_happiness
```

```python
df_happiness.info()
```

```python
df_happiness.describe()
```

# Visualization

```python
#correlation between happiness score and happniess rank is higly negative so we need to drop the happiness rank
plt.figure(figsize=(10,8))
sns.heatmap(df_happiness.corr(),annot=True)
```

```python
#checking the top 10 country which column has significant impact in happiness score
for col in df_happiness.select_dtypes(include=["float64"]):
    sns.barplot(y=df_happiness["Country"][0:10],x=df_happiness[col])
    plt.show()
```

```python
#creating boxplot to understand the idea of outliners and it indicates there are some outliners. Thus we will eliminate it by z score
df_happiness.plot(kind="box",subplots=True,layout=(3,4),figsize=(10,10))
```

```python
#dropping the columns which has no relation with the target or which has no affect on target by eliminating them
df_happiness.drop(columns=["Country","Region","Happiness Rank"],inplace=True)
```

```python
df_happiness.head()
```

```python
#simply testing the correlation after eliminating some columns reveals that all fields are required
plt.figure(figsize=(10,8))
sns.heatmap(df_happiness.corr(),annot=True)
```

```python
#only deleting the outliners if any through zscore and we can observe the structure of dataset before and after zscore applied that outliners are deleted
from scipy.stats import zscore
z_score=abs(zscore(df_happiness))
print(df_happiness.shape)
happiness=df_happiness[(z_score<3).all(axis=1)]
print(happiness.shape)
```

```python
happiness.head()
```

```python
#data division into x (independent variable) and y (target variable)
x=happiness.drop("Happiness Score",axis=1)
y=happiness["Happiness Score"]
```

```python
x
```

```python
y.head()
```

## Using MinMax Scaling technique to scale the data of x(independent variables) so that they remain in same scale

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
#data after scaling
x
```

# Training the model

```python
#importing another library for training model
from sklearn.model_selection import train_test_split
```

```python
#data is divided into train and test, with test given 20% of the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

# Using the Regression models as the target variable is continious

# LinearRegression model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#examining the result and error
print("Score",r2_score(y_test,pred))
print("mean_absolute_error",mean_absolute_error(y_test,pred))
print("mean_squared_error",mean_squared_error(y_test,pred))
```

# SVM model

```python
#determining the best-performing kernel using the SVM approach
from sklearn.svm import SVR
kernellist=['linear','poly','rbf']
for i in kernellist:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print(sv.score(x_train,y_train))
```

# Regularization :: L1 & L2 Regularizatiom
L1---Lasso Regression,
L2---Ridge Regression

```python
#using the regularisation to prevent overfitting as the score approaches 1,
#if any utilising this to gain highest score by lowering or eliminating the coefficients 
from sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
ls.coef_
```

```python
plt.figure(figsize=(18,5))
plt.bar(['Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'],ls.coef_)
plt.show()
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
plt.figure(figsize=(18,5))
plt.bar(['Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'],rd.coef_)
plt.show()
```

```python
#Since both lasso and ridge are used by elasticnet, doing this to check the result
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.0001)
enr.fit(x_train,y_train)
enrpred=enr.predict(x_test)
print(enr.score(x_train,y_train))
enr.coef_
```

# Ensemble Techniques

### Now using ensemble algorithm to reduce bias and errors and provide stability

```python
#Randomforest as bagging technique
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
```

```python
#AdaBoost as boosting technique
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

## Random forest is giving good score. linear and svm model also giving score more than random forest but just to avoid overfitting and stable model we go with random forest

### So we choose Randomforest as a model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
```
